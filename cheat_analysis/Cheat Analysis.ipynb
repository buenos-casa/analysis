```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cheat_analysis.csv')
df = df.dropna()
```

```python
df.head()
```

```python
cheat_df = df[df['barrio'] != df['place_name']]
```

```python
cheat_df.head()
```

```python
cheat_df.reset_index(inplace=True)
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('NUNEZ', 'NUÑEZ'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('NU?ÆEZ', 'NUÑEZ'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('NU?EZ', 'NUÑEZ'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('GOGHLAND', 'COGHLAN'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('PARQUE AVELLANED', 'PARQUE AVELLANEDA'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('SARSFIELD', 'SARFIELD'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('VILLA GRAL. MITRE', 'VILLA GENERAL MITRE'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('VILLA GRAL. MITR', 'VILLA GENERAL MITRE'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('VILLA GRAL MITRE', 'VILLA GENERAL MITRE'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('VILLA PUYRREDON', 'VILLA PUEYRREDON'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('PARQUE AVELLANEDAA', 'PARQUE AVELLANEDA'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('VERSALLES', 'VERSAILLES'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('MONSERRAT', 'MONTSERRAT'))))
cheat_df['place_name'] = (cheat_df['place_name'].apply(lambda x: str(x.replace('SAN NICOLAS', 'CAPITAL FEDERAL'))))
```

```python
pd.Series(sorted(cheat_df.place_name.unique()))
```

```python
cheat_df.sort_values(['barrio', 'place_name'])
```

```python
barrios_df = pd.read_csv('../../data/barrio_table.csv')
```

```python
barrio_strings = barrios_df['Barrio'].values
barrios_df.head()
```

```python
'PALERMO' in barrio_strings
```

```python
cheat_df['true_mismatch'] = [i in barrio_strings for i in cheat_df.place_name]
```

```python
cheat_df[cheat_df.true_mismatch == True].count()
```

```python
cheat_df = cheat_df[cheat_df.true_mismatch]
```

```python
cheat_df.reset_index(inplace=True)
```

```python
cheat_df = cheat_df.drop(['level_0', 'index', 'Unnamed: 0', 'barrio', 'true_mismatch'], axis=1)
```

```python
cheat_df.head()
```

```python
cheat_df['c_b_id'] = [int(barrios_df[barrios_df.Barrio == i].id) for i in cheat_df.place_name]
cheat_df = cheat_df.drop('place_name', axis=1)
```

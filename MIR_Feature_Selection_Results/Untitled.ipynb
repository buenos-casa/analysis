```python
import pandas as pd
import os
```

```python
tfifteen = []

for i in range(2015, 2019):
    directory = os.fsencode(str(i))

    for file in os.listdir(directory):
         filename = os.fsdecode(file)
         if filename.endswith(".csv"):
            print(filename)
            tfifteen.append(pd.read_csv(str(i) + "/" + filename, header=None))
            tfifteen[-1]['Year'] = i
            tfifteen[-1]['b_id'] = int(filename[5:7])
             # print(os.path.join(directory, filename))
```

```python
tfifteen[5].head()
```

```python
total = pd.concat(tfifteen, ignore_index=True)
```

```python
total.to_csv("importance.csv")
```
